```
from transformers import pipeline
```

```
def detect_mood(text):
    sentiment_analyzer = pipeline("text-classification", model="mrm8488/distilroberta-finetuned-financial-news-sentiment-analysis")
    result = sentiment_analyzer(text)
    print(result[0])
    if result[0]['label'] == 'positive':
        return 'Positive'
    elif result[0]['label'] == 'positive':
        return 'Sad'
    else:
        return 'Neutral'
detect_mood("I'm so happy!")
```

```
%%writefile app.py
import streamlit as st
from transformers import pipeline

# Define a function to detect mood
@st.cache_data
def detect_mood(text):
    sentiment_analyzer = pipeline("text-classification", model="mrm8488/distilroberta-finetuned-financial-news-sentiment-analysis")
    result = sentiment_analyzer(text)
    st.write(result)
    if result[0]['label'] == 'positive':
        return 'Happy'
    elif result[0]['label'] == 'negative':
        return 'Sad'
    else:
        return 'Neutral'

# Streamlit app
st.title('Song Recommender')

# User input
user_input = st.text_area('Enter a song title or lyrics to get recommendations:')

# Detect mood
if st.button('Get Recommendations'):
    if user_input:
        mood = detect_mood(user_input)
        st.write(f'Detected Mood: {mood}')

        # Placeholder for song recommendations
        st.write('Song Recommendations:')
        st.write('1. Happy Song')
        st.write('2. Sad Song')
        st.write('3. Neutral Melody')
        st.write('4. Upbeat Tune')
        st.write('5. Melancholic Ballad')
    else:
        st.write('Please enter a song title or lyrics to get recommendations.')
```
